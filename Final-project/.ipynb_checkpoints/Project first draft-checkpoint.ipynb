Below is the details of columns:

**isbn13** => International Standard book number with 13 digits ;
**isbn10** => International standard book number wit 10 digit ;
**title** => title of the book ;
**subtitle** => ... ;
**author** => book's writer ;
**categories** => book's type ;
**thumbnail** => book's picture.png url ;
**description** => kinda like the synopsis of the book ;
**published_year** => the year when the book published ;
**average_rating** => average rating ;
**num_page** => pages of the book ;
**rating_count** => number of ratings made for the book


**First** , import pandas as pd and numpy as np

```python
import pandas as pd
import numpy as np
```

Read the data sets using pd.read_csv() and look at the first 10 rows

```python
books = pd.read_csv('books.csv')
books.head(10)
```

Now check the shape of the data

```python
books.shape
```

We want to remove unnecssary columns from the datasets

We can use .drop(columns[]) function for this

```python
#remove isbn10, considered repeitive with isbn13
#remove the subtitle column, lots of null and not quite needed
#remove thumbnail column, only showing picture of book's front cover 
books = books.drop(columns=['isbn10','subtitle','thumbnail'])
books.head()
```

We don't want any duplication of the book in our recommendation app. So, lets check how many repetitive tiles does we have using the .total_counts()

```python
books['title'].value_counts()
```

Now that we know few books has been repeated quite a lot, we might want to drop any duplication of title beacuse our app don't want to give a result of the same books to our user.

```python
#using the drop_duplicates
books.drop_duplicates(subset = 'title',keep = False,inplace=True)
books['title'].value_counts()
```

```python
books.tail(10)
```

Let's change float published_year, num_pages and ratings_count into integer

```python
books['published_year'] = books['published_year'].fillna(0).astype(int)
```

```python
books['num_pages'] = books['num_pages'].fillna(0).astype(int)
```

```python
books['ratings_count'] = books['ratings_count'].fillna(0).astype(int)
```
